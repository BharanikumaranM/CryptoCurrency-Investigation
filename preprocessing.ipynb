```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
file_loc = "./dataset.xlsx"
```

```python
df = pd.read_excel(file_loc,usecols='A:B')
```

```python
df
```

```python
# Function to count the number of numbers and alphabets in a string
def count_numbers_and_alphabets(s):
    num_count = sum(c.isdigit() for c in s)
    alpha_count = sum(c.isalpha() for c in s)
    return num_count, alpha_count

# Apply the function to count numbers and alphabets for each row
df[['Num_Count_Private', 'Alpha_Count_Private']] = df['Private_Key'].apply(lambda x: pd.Series(count_numbers_and_alphabets(x)))
df[['Num_Count_Public', 'Alpha_Count_Public']] = df['Public_Key'].apply(lambda x: pd.Series(count_numbers_and_alphabets(x)))

```

```python
df = df.drop('Cosine_Similarity', axis=1)
df
```
